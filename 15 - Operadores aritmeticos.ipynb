# Operadores aritméticos

Para realizar operaciones aritméticas en Python necesitamos dos valores, que pueden ser números enteros, flotantes o variables que contengan estos valores.


## Valores base


```python
a = 10
b = 3
```

## Suma


```python
print(a + b)
```

## Resta


```python
print(a - b)
```

## Multiplicación


```python
print(a * b)
```

## División


```python
print(a / b)
```

## División entera


```python
print(a // b)
```

## Módulo


```python
print(a % b)
```

## Potenciación


```python
print(a ** b)
```

## Orden de operaciones (PEMDAS)

El orden de las operaciones aritméticas en Python sigue las reglas de PEMDAS.

- Paréntesis
- Exponentes
- Multiplicación
- División
- Adición
- Sustracción

Esto significa que las operaciones dentro de paréntesis se realizan primero, seguidas por exponentes, luego multiplicación y división, y finalmente adición y sustracción (ambas leídas de izquierda a derecha).


```python
print((a + b) * (a - b) / b)
```

## Funciones con operadores aritméticos

Podemos usar operadores aritméticos para crear funciones que realicen cálculos específicos.
Por ejemplo, podemos crear una función para verificar si un número es impar o par.


### Número impar


```python
def is_odd(number):
    return number % 2 != 0

print(is_odd(4))
```

### Número par


```python
def is_even(number):
    return number % 2 == 0

print(is_even(2))
```
