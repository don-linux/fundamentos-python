# Funciones

Una función es un bloque de código que se ejecuta cuando se le llama.

Permite modularizar el código y organizarlo en bloques reutilizables.

Una función está compuesta por:

- **Argumentos**: el código con el que trabaja la función (los nombres definidos en la función).
- **Parámetros**: el código que se envía a la función cuando se le llama.


## Definir una función

Las funciones se definen con la palabra reservada `def`, seguida del nombre de la función y sus argumentos entre paréntesis. Podemos asignar valores por defecto a los argumentos para que sean opcionales al llamar a la función.


```python
def salute(name, last_name='', nationality='world'):
    print('Hi', name, last_name, 'from', nationality)
```

## Llamar a una función

Para ejecutar una función la llamamos por su nombre y le pasamos los parámetros entre paréntesis. Si la función tiene argumentos con valores por defecto, podemos omitirlos.


```python
salute('Fer', 'Diaz', 'Mexico')
salute('Fercho')
```

## Devolver resultados con `return`

Una función puede devolver resultados usando la palabra reservada `return`. A diferencia de `print`, que solo muestra el valor en consola, `return` permite que el resultado pueda ser utilizado por otra parte del programa.


```python
def sum(a, b):
    return a + b

sum(5, 5)
```

## Asignación a una variable

El valor devuelto por una función puede asignarse a una variable para usarlo después.


```python
x = sum(2, 2)

print(x)
```

## Función vacía con `pass`

Si declaramos una función pero aún no queremos definir su contenido, podemos usar `pass` para evitar un error de sintaxis.


```python
def myfunc():
    pass
```
