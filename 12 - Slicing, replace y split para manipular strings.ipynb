# Manipulación de strings en Python


## Acceso por índice

Podemos acceder a caracteres individuales en un string usando índices.
Los índices empiezan en 0.


```python
text = "Python is dope"
print(text[5])
```

## Slicing

Podemos obtener subcadenas usando slicing.
Estas tienen rangos de inicio a fin, donde el fin no está incluido.
Siempre debemos añadir un índice extra al final.


```python
print(text[0:6])
```

## Replace

Con `replace` podemos reemplazar contenido dentro de un string si existe el contenido a reemplazar.


```python
content_course = "This course is about C#"
print(content_course.replace("C#", "Python"))
```

## Split

Con `split` podemos dividir un string en una lista de strings usando un separador específico.


```python
course_splited = content_course.split(" ")
print(course_splited)
```

## Normalización sin distinguir mayúsculas y minúsculas

Podemos normalizar strings para hacer comparaciones sin importar mayúsculas o minúsculas.


```python
text = "HelLo WorLD"
wanted_word = "world"

print(wanted_word in text)
print(wanted_word.lower() in text.lower())
```
