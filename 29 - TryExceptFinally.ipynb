# Try, Except y Finally

Los errores en Python siempre ocurrirán y, como en todos los lenguajes, detienen la ejecución del programa.

Para esto fueron creados `try` y `except`: para manejar los errores de una mejor manera.

## Ejemplo básico

Aunque sabemos que una división por cero no puede hacerse, puede que haya más código que podamos ejecutar. Entonces le pasamos al `except` el error que esperamos (`ZeroDivisionError`), y que nos devuelva, en este caso en consola, un mensaje que nos indique el error.

Pero que permita aun así la ejecución del resto del programa. De no ponerlo, simplemente el programa se detiene.

```python
try:
    div = 10 / 0
except ZeroDivisionError:
    print("A number cannot be divided by zero")
```

## Recuperarse de un error y `finally`

Para un script corto puede no ser tan útil, pero en programas grandes o aplicaciones web que no deberían detener su ejecución tan fácilmente, es importante tener una o muchas alternativas para seguir con la ejecución del programa.

En el siguiente ejemplo el error sucede (`NameError`, porque la variable `name` no está definida), pero le ofrecemos al usuario un valor por defecto y listo: el programa falla, pero lo arreglamos en runtime.

El bloque `finally` se ejecuta siempre, sin importar si el código dentro del `try` pasó o falló.

```python
try:
    print(name)
except NameError:
    name = "Freddy"
    print("The name of the variable hasn't been defined, defaulting the variable name to 'Freddy'")
    print(name)
finally:
    print("This print will be executed always, even the code is passing or failing")
```
