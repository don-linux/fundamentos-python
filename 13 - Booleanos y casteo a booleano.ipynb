# Booleanos en Python

La forma de escribir booleanos en Python es con la primera letra en mayúscula.


## Valores booleanos


```python
T = True
F = False
```

## Comparaciones que devuelven booleanos


```python
print(3 == 3)
print(5 > 3)
print(5 > 10)
```

## Castear a booleano

Algunos valores se convierten a `True` cuando se usan con `bool()`.

- Cualquier string no vacío se convierte a `True`.
- Cualquier número distinto de 0 se convierte a `True`.
- Cualquier lista no vacía se convierte a `True`.


```python
print(bool("Hi"))
print(bool(1))
print(bool(["manzana", "pera"]))
```

## Casos que se convierten a `False`

- Cualquier string vacío se convierte a `False`.
- El número 0 se convierte a `False`.
- Las listas vacías se convierten a `False`.
- `None` se convierte a `False`.


```python
print(bool(""))
print(bool(0))
print(bool([]))
print(bool(None))
```

## Verificación de tipos con `isinstance`

Si el valor coincide con el tipo de dato, `isinstance` devuelve `True`. El valor `1` no se convierte a `True` con `isinstance`.


```python
x = 0
y = 1
z = True

print(isinstance(x, int))
print(isinstance(y, bool))
```
