# Manipular archivos

Para trabajar con archivos en Python usamos la función `open`.

**Sintaxis:**

```python
open(nombreArchivo, modo)
```

Los modos más comunes son:

- **`r`** (Read): lectura.
- **`w`** (Write): escribe o sobreescribe en archivos existentes.
- **`x`** (Execute): crea archivos nuevos; si el archivo ya existe, genera un error.

## Leer un archivo

Abrimos `file.txt` en modo lectura (`r`), leemos una línea con `readline` y cerramos el archivo con `close`.

Como abrir un archivo que no existe genera un `FileNotFoundError`, envolvemos el código en un `try-except` para manejar ese caso sin detener el programa.

````python
try:
    f = open("file.txt", "r")
    print(f.readline())
    f.close()
except FileNotFoundError:
    print("File not found")
````
