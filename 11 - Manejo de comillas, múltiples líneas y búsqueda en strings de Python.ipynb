# Manejo de comillas, múltiples líneas y búsqueda en strings de Python


## Comillas dentro de comillas


```python
print("Hola 'mundo'")
print('Hola "mundo"')
```

## Múltiples líneas con comillas triples


```python
multiples = """Hola
Mundo"""
print(multiples)
```

## `in` y `not in`: búsqueda en strings


```python
texto = "Este curso es de fundamentos de Python"

print("Python" in texto)
print("python" in texto)
print("JavaScript" not in texto)
```

## `.upper` y `.lower`: convertir a mayúsculas y minúsculas


```python
mayuscula = texto.upper()
minuscula = texto.lower()

print(mayuscula)
print(minuscula)
```

## `.strip`: eliminar espacios al inicio y al final


```python
espacios = "   este es el texto   "
sin_espacios = espacios.strip()

print(sin_espacios)
```
