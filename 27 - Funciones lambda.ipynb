# Funciones lambda

Una función `lambda` es una función pequeña y anónima.

Puede tener muchos argumentos, pero solo una expresión.

**Sintaxis:**

```python
lambda argumentos: expresion
```


## Ejemplo básico

Asignamos una función lambda a la variable `x`. La función toma un argumento `a` y devuelve `a + 10`.

Al llamar `x(5)`, el `print` toma la variable `x` (que contiene la función lambda) y le pasa el argumento `5`, el cual es el valor que toma `a`. Entonces `a + 10` se convierte en `5 + 10 = 15` y el resultado se imprime.


````python
x = lambda a : a + 10

print(x(5))
````

## Múltiples argumentos

Una función lambda puede recibir varios argumentos separados por comas. En este caso, el `print` pasa los argumentos `5` y `6` a la función lambda, que multiplica ambos valores.


````python
x = lambda a, b : a * b

print(x(5, 6))
````

## Fábrica de funciones

Las funciones lambda son especialmente útiles cuando se usan dentro de otras funciones, permitiéndonos crear funciones personalizadas a partir de una plantilla.

Aquí, `myfunc(n)` devuelve una nueva función lambda que multiplica su argumento por `n`. Así podemos generar funciones como `double` (multiplicar por 2) y `triple` (multiplicar por 3).


````python
def myfunc(n):
    return lambda a : a * n

double = myfunc(2)
triple = myfunc(3)

print(double(5))
print(triple(2))
````
