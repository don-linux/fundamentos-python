# Bucles for


## Recorrer un string


```python
palabra = 'Python'

for letra in palabra:
    print(letra)
```

## Break en un bucle `for`

Python es sensible a mayúsculas y minúsculas al comparar los valores.


```python
for letra in palabra:
    if letra == 'y':
        break
    print(letra)
```

## Recorrer una lista


```python
frutas = ['Manzana', 'Banana', 'Cereza']

for fruta in frutas:
    print(fruta)
```

## Continue y else


```python
for fruta in frutas:
    if fruta == 'Banana':
        continue
    print(fruta)
else:
    print('Bucle finalizado')
```

## Range

En Python los rangos no incluyen el valor final, por lo que `range(1, 6)` genera los números del 1 al 5.


```python
print('-------------------------')

for i in range(0,10,2):
    print(i)
```

## Bucles anidados


```python
models = ['ChatGPT', 'Claude', 'Gemini']
apps = ['is awesome', 'is cool']

for model in models:
    for app in apps:
        print(model, app)
```

## Orden de los bucles anidados


```python
frutas = ["manzana", "naranja", "kiwi"]
adjetivos = ["rica", "saludable"]

for adjetivo in adjetivos:
    for fruta in frutas:
        print(fruta, adjetivo)
```

Al invertirlo, primero el bucle de frutas pasa por el bucle de adjetivos.


```python
frutas = ["manzana", "naranja", "kiwi"]
adjetivos = ["rica", "saludable"]

for fruta in frutas:
    for adjetivo in adjetivos:
        print(fruta, adjetivo)
```
