# Listas

Las listas en Python son colecciones ordenadas, que permiten mutabilidad, además de contener elementos duplicados y de diferentes tipos de datos.

Pueden contener elementos duplicados al igual que las tuplas.


## Crear una lista


```python
tesla_models = ['Model S', 'Model 3', 'Model X', 'Model Y', 'CyberTruck']

print(tesla_models)
```

## Acceder a elementos por índice

Para acceder a un elemento de la lista se utiliza el índice.


```python
print(tesla_models[4])
```

## Mutabilidad

Las listas son mutables, lo que significa que se pueden modificar después de su creación.


```python
tesla_models[0] = 'S'
tesla_models[1] = 3
tesla_models[2] = 'X'
tesla_models[3] = 'Y'

print('Tesla is', tesla_models)
```

## Listas con diferentes tipos de datos

Una lista puede contener diferentes tipos de datos como enteros, strings, decimales, booleanos, tuplas, sets y diccionarios. Además, pueden ser modificados después de su creación.


```python
mixed_list = [1, "hola", 3.14, True, (1, 2, 3), {1, 2, 3}, {'clave': 'valor'}]

print(mixed_list)
print(type(mixed_list))
```

## Longitud de una lista

Para saber la longitud de una lista se utiliza la función `len()`.


```python
print(len(mixed_list))
```

## Rangos de índices en una lista

En Python los rangos no incluyen el valor final, por lo que `1:3` imprime los elementos del 1 al 2.


```python
print(tesla_models[1:3])
```

### Desde el inicio

Imprime desde el inicio hasta el índice 2. El índice 3 no está incluido.


```python
print(tesla_models[:3])
```

### Hasta el final

Imprime desde el índice 1 hasta el final.


```python
print(tesla_models[1:])
```

## Buscar un elemento en una lista

Mediante una condicional `if` se puede verificar si un elemento está presente en la lista utilizando el operador `in`.


```python
if 'X' in tesla_models:
    print('Model X available!')
```

## Iterar y guardar elementos en una lista con `for`


```python
lista = []

for i in range(1,6):
    lista.append(i)

print(lista)
```

## Métodos


```python
frutas = ['manzana', 'banana', 'naranja']
```

### Append

Agrega un elemento al final de la lista.


```python
frutas.append('pera')
print(frutas)
```

### Insert

Permite insertar en la posición deseada. No elimina el elemento que se encuentra en esa posición, sino que lo desplaza una posición a la derecha.

El primer parámetro es el índice y el segundo el elemento a insertar.


```python
frutas.insert(1, 'uva')
print(frutas)
```

### Remove

Elimina el elemento declarado, si existe.


```python
frutas.remove('manzana')
print(frutas)
```

### Pop

Permite eliminar el elemento declarando el índice.


```python
frutas.pop(3)
print(frutas)
```

### Sort

Sirve para ordenar alfabéticamente, de menor a mayor, o de mayor a menor dependiendo del tipo de datos que contenga la lista.


```python
frutas.sort()
print(frutas)
```

### Reverse

Sirve para ordenar a la inversa.


```python
frutas.reverse()
print(frutas)
```

## Unir listas


```python
new_list = tesla_models + frutas
print(new_list)
```
