# Módulos

Un módulo es un archivo, externo a nuestro programa, que contiene funciones y variables que podemos utilizar en nuestro programa.

Por ejemplo, podemos tener un módulo que contenga funciones para manejar archivos, otro para manejar fechas, otro para manejar números, etc.

Y estos podemos importarlos a nuestro script o programa principal.

## Crear un módulo

Supongamos que tenemos un archivo llamado `ops.py`, que funciona como módulo y contiene una función para sumar dos números.

```python
# ops.py

def addition(a, b):
    return a + b
```

## Importar un módulo

Desde nuestro programa principal (por ejemplo `main.py`), podemos importar la función `addition` del módulo `ops` y usarla como `add`.

Con `as`, podemos asignar un nombre alternativo a la función o variable que importamos. Si lo omitimos, el nombre de la función o variable será el mismo que tiene en el módulo.

Y así con N cantidad de módulos y funciones que podemos importar a nuestro programa principal.

> **Nota:** para que este ejemplo funcione, el archivo `ops.py` debe existir en el mismo directorio que el programa principal.

```python
# main.py

from ops import addition as add

print(add(1, 2))
```
