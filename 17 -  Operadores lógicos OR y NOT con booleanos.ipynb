# Operadores de comparación y operadores lógicos

Estos operadores nos permiten comparar valores y devolver un valor booleano (`True` o `False`), los cuales nos indican si la comparación es verdadera o falsa.


## Valores base


```python
x = 5
y = 3
```

## Igual a


```python
print(x == y)
```

## Diferente de


```python
print(x != y)
```

## Mayor que


```python
print(x > y)
```

## Menor que


```python
print(x < y)
```

## Mayor o igual que


```python
print(x >= y)
```

## Menor o igual que


```python
print(x <= y)
```

## Operadores lógicos `and`, `or` y `not`

Estos operadores lógicos nos permiten combinar múltiples condiciones booleanas y devolver un valor booleano (`True` o `False`) basado en la evaluación de esas condiciones.


### `and`

Verdadero si ambas condiciones son verdaderas.


```python
print(x > 2 and y < 5)
```

### `or`

Verdadero si alguna de las condiciones es verdadera.


```python
print(x > 2 or y > 5)
```

### `not`

Invierte el valor booleano de la condición.


```python
print(not(x > 2))
```
