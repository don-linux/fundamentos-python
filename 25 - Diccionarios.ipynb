# Diccionarios

Un diccionario es una colección de pares clave-valor.
A partir de Python 3.7, los diccionarios mantienen el orden de inserción.


## Crear un diccionario


```python
cafe = {
    "grano": "geisha",
    "pais": "colombia",
    "año": 2025,
}

print(cafe)
```

## Buscar una clave en el diccionario


```python
if "grano" in cafe:
    print('Si hay!')
```

## Recorrer un diccionario


### Recorrer claves


```python
for k in cafe:
    print(k)
```

### Recorrer valores


```python
for v in cafe.values():
    print(v)
```

### Recorrer claves y valores


```python
for k, v in cafe.items():
    print(k, v)
```

## Acceder a valores

Podemos acceder a los valores del diccionario usando corchetes para indicar la clave y obtener el valor.


```python
print(cafe["pais"])
```

## Acceder con `get()`


```python
print(cafe.get("grano"))
```

## Obtener las claves


```python
print(cafe.keys())
```

## Obtener los valores


```python
print(cafe.values())
```

## Agregar un nuevo par clave-valor


```python
cafe["tostado"] = "Suave"
print(cafe)
```

## Actualizar un par clave-valor

Si la clave ya existe, podemos actualizar su valor asignando un nuevo dato.


```python
cafe["año"] = 2020
print(cafe)
```

## Actualizar o agregar con `update()`


```python
cafe.update({"año": 2022, "grano": "arabica"})
```

## Eliminar un par clave-valor


```python
cafe.pop("año")
print(cafe)
```

## Eliminar el último par clave-valor


```python
cafe.popitem()
print(cafe)
```

## Eliminar todos los pares clave-valor


```python
cafe.clear()
print(cafe)
```

## Eliminar el diccionario completo


```python
del cafe
print(cafe)
```

## Diccionario anidado


```python
autos = {
    "tesla": {
        "model": "X"
    },
    "cybertruck": {
        "model": "Founders"
    }
}

print("Model", autos["tesla"]["model"])
```
