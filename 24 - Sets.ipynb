# Sets (conjuntos) en Python

Los sets son colecciones no ordenadas de elementos únicos. Se definen utilizando llaves `{}` o la función `set()`. Son mutables, no permiten elementos duplicados y no se pueden acceder por índice.

Si existen elementos duplicados no arrojan un error, simplemente se ignoran y no se agregan al set.


## Crear un set

La longitud del set es 3, ya que `'manzana'` se repite y se ignora.


```python
frutas = {'manzana', 'banana', 'naranja', 'manzana'}

print(frutas)
print(type(frutas))
print(len(frutas))
```

## Tipos de datos dentro de un set

Un set puede contener diferentes tipos de datos inmutables (hashables), como enteros, strings, decimales, booleanos, tuplas (con elementos hashables) y frozensets. Además, puede ser modificado después de su creación.

No pueden contener listas, diccionarios u otros sets, ya que estos no son hashables (son mutables o no tienen hash definido).


```python
mixed_set = {1, "hola", 3.14, True}

print(mixed_set)
print(type(mixed_set))
```

## Recorrer un set


```python
for item in mixed_set:
    print(item)
```

## Buscar elementos en un set


```python
if 1 in mixed_set:
    print('Founded!')
```

## Métodos en un set


### Crear un set vacío


```python
aws_services = set()
```

### Add


```python
aws_services.add('Route 53')
print(aws_services)
```

### Update

Al actualizar podemos agregar el contenido de otras listas, tuplas o conjuntos.


```python
new_services = {'EC2', 'ECC', 'S3'}

aws_services.update(new_services)
print(aws_services)
```

### Remove

`remove` comprueba que el elemento a remover exista. Si no existe, lanzará un error.


```python
aws_services.remove('ECC')
print(aws_services)
```

### Discard

`discard`, si no encuentra el elemento a eliminar, simplemente mostrará los elementos restantes.


```python
aws_services.discard('ECC')
print(aws_services)
```

### Pop

`pop` elimina un elemento aleatorio del set y lo devuelve. Si el set está vacío, lanzará un error.


```python
aws_services.pop()
print(aws_services)
```

### Clear

`clear` elimina todos los elementos del set, dejando un set vacío.


```python
aws_services.clear()
print(aws_services)
```

## Operaciones de conjuntos


```python
a = {'a', 'b', 'c'}
b = {'c', 'd', 'e'}
```

### Union

Es la creación de un nuevo set, uniendo el set principal y el set a evaluar.


```python
c = a.union(b)
print(c)
```

### Intersección

Son los elementos que coinciden en el set a evaluar.


```python
i = a.intersection(b)
print(i)
```

### Diferencia

Solo los elementos que no están presentes en el set a evaluar.


```python
diff = a.difference(b)
print(diff)
```
