# Variables en Python

Una variable es un espacio en memoria donde guardamos un valor para usarlo después.
Con el signo igual (`=`), que es el operador de asignación, definimos el nombre y asignamos el valor.

## Consideraciones importantes

- Con `print` no se imprime el nombre de la variable, sino su valor.
- Python permite sobreescribir variables según el orden del código.
- Python es case sensitive: variables con el mismo nombre pero en mayúsculas y minúsculas serán diferentes.


## Asignación básica


```python
x = "New variable"
print(x)  # Imprime: New variable
```

## Reasignación de variables


```python
x = "Overwriting x"
print(x)  # Imprime: Overwriting x
```

## Sensibilidad a mayúsculas y minúsculas

Ambas variables imprimen valores diferentes.


```python
x = "lowercase"
X = "uppercase"

print(x)  # lowercase
print(X)  # uppercase
```
